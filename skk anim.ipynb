```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML

import random
```

```python
max_x = 40
max_y = 40
fig, ax = plt.subplots()
ax.set_xlim(0, max_x)
ax.set_ylim(0, max_y)
s1 = None
s2 = None
print(id(s1), id(s2))
```

```python
sx = [1, 2, 3]
sy = [4, 3, 2]
s = plt.plot(sx, sy)
```

```python
def init():
    global s1, s2
    print("returning!")
    #return s1, s2
    
    normal_p = np.empty([1, 2])
    inf_p = np.empty([1, 2])
    #print(id(s1), id(s2))
    for i in range(normal_p.shape[0]):
        normal_p[i][0] = random.randint(0,max_x)
        normal_p[i][1] = random.randint(0,max_y)
    for i in range(inf_p.shape[0]):
        inf_p[i][0] = random.randint(0,max_x)
        inf_p[i][1] = random.randint(0,max_y)
    
    #s1 = ax.scatter([],[])
    #s2 = ax.scatter([],[])
    
    if s1 is not None:
        print("init s1 not None: ", s1.get_offsets(), ", removing...")
        s1.remove()
    if s2 is not None:
        print("init s2 not None: ", s2.get_offsets(), ", removing...")
        s2.remove()
    s1 = ax.scatter(normal_p[:,0], normal_p[:,1], color = 'green')
    s2 = ax.scatter(inf_p[:,0], inf_p[:,1], color = 'red')
    #s1.set_offsets(normal_p)
    #s2.set_offsets(inf_p)
    #print("init done! ", s1, s2)
    print("s1 is : ", s1.get_offsets())
    print("s2 is : ", s2.get_offsets())
    
    return s1, s2
```

```python
def animate(i):
    global s1, s2
    if i == 0:
        
        normal_p = np.empty([1, 2])
        inf_p = np.empty([1, 2])
        #print(id(s1), id(s2))
        for i in range(normal_p.shape[0]):
            normal_p[i][0] = random.randint(0,max_x)
            normal_p[i][1] = random.randint(0,max_y)
        for i in range(inf_p.shape[0]):
            inf_p[i][0] = random.randint(0,max_x)
            inf_p[i][1] = random.randint(0,max_y)

        #s1 = ax.scatter([],[])
        #s2 = ax.scatter([],[])

        if s1 is not None:
            print("init s1 not None: ", s1.get_offsets(), ", removing...")
            s1.remove()
        if s2 is not None:
            print("init s2 not None: ", s2.get_offsets(), ", removing...")
            s2.remove()
        s1 = ax.scatter(normal_p[:,0], normal_p[:,1], color = 'green')
        s2 = ax.scatter(inf_p[:,0], inf_p[:,1], color = 'red')
        #s1.set_offsets(normal_p)
        #s2.set_offsets(inf_p)
        #print("init done! ", s1, s2)
        print("s1 is : ", s1.get_offsets())
        print("s2 is : ", s2.get_offsets())
        return s1, s2
        
    #print("Animate: ", i, id(s1), id(s2))
    #if i == 0:
    #    print(s1.get_offsets())
    
    norm_p = s1.get_offsets()
    #print("got s1 ", norm_p)
    for j in range(norm_p.shape[0]):
        tmp = norm_p[j][0] 
       # print("updating ", i, " from ", norm_p[i][0], " to ", norm_p[i][0] + 0.1 * tmp)
        norm_p[j][0] = norm_p[j][0] + 0.1 * tmp
        if norm_p[j][0] > max_x:
            norm_p[j][0] = norm_p[j][0] - max_x
        tmp = norm_p[j][1] 
        #norm_p[j][1] = norm_p[j][1] + 0.1 * tmp
        if norm_p[j][1] > max_y:
            norm_p[j][1] = norm_p[j][1] - max_y
    s1.remove()
    s1 = ax.scatter(norm_p[:,0], norm_p[:,1], color='green')
    #s1.set_offsets(norm_p)
    #if i == 0:
    #print("A ", i, " s1 set to ", norm_p)
    infe_p = s2.get_offsets()
    for j in range(infe_p.shape[0]):
        tmp = infe_p[j][0] 
        #infe_p[j][0] = infe_p[j][0] + 0.1 * tmp
        if infe_p[j][0] > max_x:
            infe_p[j][0] = infe_p[j][0] - max_x
        tmp = infe_p[j][1] 
        infe_p[j][1] = infe_p[j][1] + 0.1 * tmp
        if infe_p[j][1] > max_y:
            infe_p[j][1] = infe_p[j][1] - max_y
    s2.remove()
    s2 = ax.scatter(infe_p[:,0], infe_p[:,1], color='red')
    #s2.set_offsets(infe_p)
    #if i == 0:
    #print("A ", i, " s2 set to ", infe_p)
    return s1, s2
```

```python
ax.clear()
anim = animation.FuncAnimation(fig, animate, init_func=init, interval=100, blit = True, save_count=100)
plt.show()
```
